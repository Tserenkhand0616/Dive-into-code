[Problem 1] Creating random numbers
Create 500 random numbers with a two-dimensional normal distribution whose mean is (-3, 0) and whose covariance matrix is

[
1.0
0.8
0.8
1.0
]
Multidimensional normally distributed random numbers can be created by using np.random.multivariate_normal() .

```python
# Creating random data
import numpy as np
import matplotlib.pyplot as plt
mean = (-3, 0)
cov = [[1.0, 0.8], [0.8, 1.0]]
np.random.seed(0)
random_numbers = np.random.multivariate_normal(mean, cov, size=500)
```

```python
import numpy as np
import matplotlib.pyplot as plt

mean = [-3, 0]
cov = [[1.0, 0.8], [0.8, 1.0]]
random_numbers = np.random.multivariate_normal(mean, cov, 500)

# Create two histograms for each dimension
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.hist(random_numbers[:, 0], bins=20, alpha=0.5)
plt.title("Histogram of first row ")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.xlim(-6, 4) # Set the range of the horizontal axis

plt.subplot(1, 2, 2)
plt.hist(random_numbers[:, 1], bins=20, alpha=0.5)
plt.title("Histogram of second row")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.xlim(-6, 4) # Set the range of the horizontal axis

plt.show()
```
